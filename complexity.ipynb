```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
mainPath = r"X:\Thesis\complexityMinutes.xlsx"

complexDF = pd.read_excel(mainPath)
```

```python
complexDF
```

```python
ascendingDF = complexDF.loc[complexDF['Sequence'] == 1].copy()
seq1 = complexDF.loc[complexDF['Sequence'] == 2].copy()
seq2 = complexDF.loc[complexDF['Sequence'] == 3].copy()
seq3 = complexDF.loc[complexDF['Sequence'] == 4].copy()

```

```python
sns.set_theme()
sns.set_style('ticks', {'axes.grid': True})
sns.set_context("paper")
fig, ax1 = plt.subplots(figsize=(6, 5))
sns.lineplot(x=ascendingDF['Minute'], y=ascendingDF['Complexity'], linewidth=3)
ax1.set_xlabel('Time (min.)', fontsize=15)
ax1.set_ylabel('Complexity', fontsize=15)
# sns.despine()
plt.tight_layout()
plt.savefig(f'ascending_compelxity.pdf', dpi = 300)
```

```python
sns.set_theme()
sns.set_style('ticks', {'axes.grid': True})
sns.set_context("paper")
fig, ax1 = plt.subplots(figsize=(6, 5))
sns.lineplot(x=seq1['Minute'], y=seq1['Complexity'], linewidth=3)
ax1.set_xlabel('Time (min.)', fontsize=15)
ax1.set_ylabel('Complexity', fontsize=15)
plt.tight_layout()
plt.savefig(f'sequence1.pdf', dpi = 300)
```

```python
sns.set_theme()
sns.set_style('ticks', {'axes.grid': True})
sns.set_context("paper")
fig, ax1 = plt.subplots(figsize=(6, 5))
sns.lineplot(x='Minute', y='Complexity', data=seq2, linewidth=3)
ax1.set_xlabel('Time (min.)', fontsize=15)
ax1.set_ylabel('Complexity', fontsize=15)
plt.tight_layout()
plt.savefig(f'sequence2.pdf', dpi = 300)
```

```python
sns.set_theme()
sns.set_style('ticks', {'axes.grid': True})
sns.set_context("paper")
fig, ax1 = plt.subplots(figsize=(6, 5))
sns.lineplot(x='Minute', y='Complexity', data=seq3, linewidth=3)
ax1.set_xlabel('Time (min.)', fontsize=15)
ax1.set_ylabel('Complexity', fontsize=15)
plt.tight_layout()
plt.savefig(f'sequence3.pdf', dpi = 300)
```
